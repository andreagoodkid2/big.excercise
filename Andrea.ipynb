```python
#BAI 1

def search(arr, n, x):
	for i in range(0, n):
		if (arr[i] == x):
			return i
	return -1

arr = [1,2,3,4,5,6,7,8,9]
x = int(input("Nhap ma so sinh vien: "))
n = len(arr)
result = search(arr, n, x)

if(result == -1):
	print("Sinh vien khong co trong lop")
else:
	print("Sinh vien co mat trong lop")
    
print("So luong sinh vien hien tai co trong lop:", n)
```
